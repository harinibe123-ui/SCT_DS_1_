```python
import matplotlib.pyplot as plt
import numpy as np

# Sample continuous
ages = np.random.normal(30, 12, 1000)
plt.figure(figsize=(8,5))
plt.hist(ages, bins=10)  # number of bars (bins)
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```
